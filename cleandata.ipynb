```python
import pandas as pd
import re

file_path = "tcas_engineering_programs.xlsx"
data = pd.read_excel(file_path)
```

```python
data = data[['ลำดับ', 'คำค้นหา', 'มหาวิทยาลัย', 'หลักสูตร', 'ประเภทหลักสูตร', 'ค่าใช้จ่าย', 'URL', 'ข้อมูลเริ่มต้น']]
data.head()
```

```python
data = data.drop(columns=['ข้อมูลเริ่มต้น'])
data = data.dropna(subset=['มหาวิทยาลัย'])
data = data.drop_duplicates(subset=['มหาวิทยาลัย', 'หลักสูตร'])
# data.tail()  
```

```python
data['มหาวิทยาลัย'] = data['มหาวิทยาลัย'].str.replace('ข้อมูลการรับสมัครของ', '', regex=False)
data.head()
```

```python
data.info() 
```

```python
# Filter rows where 'หลักสูตร' contains either 'วิศวกรรมศาสตร์' or 'วศ.บ.'
data = data[data['หลักสูตร'].str.contains('วิศวกรรมคอมพิวเตอร์|วิศวกรรมศาสตรบัณฑิต|วิศวกรรมศาสตร์|วศ.บ.', na=False)]

# Show the number of rows after filtering
print(f"Number of rows after filtering: {len(data)}")
data.head()
```

```python
def extract_major(course_name):
    # ลบคำที่ไม่เกี่ยวข้อง
    course_name = re.sub(r'(วิศวกรรมศาสตรบัณฑิต|หลักสูตร|วศ\.บ\.|วิทยาเขต|ภาษาไทย|นานาชาติ|ปกติ|Joint Degree - )', '', course_name)
    # ค้นหาข้อความที่เริ่มต้นด้วย "วิศวกรรม" และจบที่เว้นวรรค (หรือ "และ" ตามด้วยคำถัดไป)
    match = re.search(r'(วิศวกรรม[^\s]*)(?:\sและ\s[^\s]*)?', course_name)
    return match.group(0) if match else None

# ใช้ฟังก์ชันกับคอลัมน์ 'หลักสูตร' และลบวงเล็บปิดถ้ามี
data['ชื่อสาขาวิชา'] = data['หลักสูตร'].apply(extract_major).str.rstrip(')')

# ลบค่าที่ซ้ำกัน
unique_majors = data['ชื่อสาขาวิชา'].str.replace(r'\)$', '', regex=True).str.extract(r'(.+?)(?:\sและ\s.+)?$')[0]
data.head()
```

```python
cols = list(data.columns)
cols.remove('ชื่อสาขาวิชา')
new_cols = cols[:3] + ['ชื่อสาขาวิชา'] + cols[3:]
data = data[new_cols]
data.head()
```

```python
def process_cost_column(text):
    if pd.isna(text):
        return None
    if "ตลอดหลักสูตร" in text:
        # Extract all numbers and find the maximum
        numbers = re.findall(r'[\d,]+(?:\.\d+)?', text.replace(',', ''))
        if numbers:
            max_number = max(float(num) for num in numbers)
            return max_number / 4 / 2
    return extract_number(text)

def extract_number(text):
    if isinstance(text, float):  # If the value is already a float, return it
        return text
    if not isinstance(text, str):  # If the value is neither a string nor a float, return None
        return None
    # Find all numbers (with optional commas and decimals)
    numbers = re.findall(r'[\d,]+(?:\.\d+)?', text.replace(',', ''))
    if numbers:
        # Return the first number found as float
        return float(numbers[0])
    return None


data['ค่าใช้จ่าย'] = data['ค่าใช้จ่าย'].apply(extract_number)
data['ค่าใช้จ่าย'] = data['ค่าใช้จ่าย'].apply(lambda x: None if x is not None and x < 1000 else x)
data.head()
```

```python
data['ค่าใช้จ่าย'] = data['ค่าใช้จ่าย'].apply(lambda x: x / 4 / 2 if x is not None and x >= 100000 else x)
data.head()
```

```python
data['ค่าใช้จ่าย'] = data['ค่าใช้จ่าย'].fillna("ไม่ระบุบนเว็บไซต์Tcas")
data.head()
```

```python
data['ลำดับ'] = range(1, len(data) + 1)
data.head()
```
